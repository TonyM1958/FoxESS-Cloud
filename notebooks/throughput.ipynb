```python
# setup for testing
!pip install foxesscloud --root-user-action=ignore --quiet --upgrade
import foxesscloud.openapi as f
import json
import matplotlib.pyplot as plt

f.api_key = "my.fox_api_key"
```

```python
f.get_real(v=['energyThroughput'])
```

```python
y = 2024
m = 1
data = []
for i in range(0,50):
    result=f.get_history(d=f"{y:04}-{m:02}-01", v=['energyThroughput'])
    if len(result[0]['data']) > 0:
        data.append(result[0]['data'][0])
    if m == 12:
        m = 1
        y += 1
    else:
        m += 1
    if y == 2026 and m > 1:
        break
print(f"Energy Throughput:")
for d in data:
    print(f"  {d['time'][:10]} = {d['value']:5.0f} kWh")
```

```python
figure_width = 9
plt.figure(figsize=(figure_width, figure_width/3))
x = [i for i in range(0, len(data))]
y = [d['value'] for d in data]
plt.plot(x, y, label = f"Energy Throughput")
plt.xticks(ticks=x, labels=[d['time'][:10] for d in data], rotation=90, fontsize=8, ha='center')
plt.title(f"Energy Throughput by Month", fontsize=12)
plt.grid()
plt.show()
```
